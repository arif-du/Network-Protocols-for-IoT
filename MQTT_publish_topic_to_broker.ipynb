# **Publishing to a Topic (inside temperature)**

image.png

This Python script connects to an MQTT broker, generates a random temperature value between 20.0 and 21.0, publishes it to the "TEMPERATURE" topic, and then repeats the process every second. This is a basic simulation of an MQTT client that sends temperature data at regular intervals.

The first application will publish the inside temperature on this topic.

**Algorithm of publishing to a topic**

1.	Import the necessary modules: Import the paho.mqtt.client module from Paho MQTT, and the randrange and uniform functions from the random module. Also, import the time module.
2.	Set up MQTT broker and client: Set the mqttBroker variable to the MQTT broker's address, in this case, "mqtt.eclipseprojects.io". Create an MQTT client instance named "Temperature_Inside" using mqtt.Client("Temperature_Inside").
3.	Connect to the MQTT broker: Use the connect() method of the MQTT client to connect to the specified MQTT broker.
4.	Start an infinite loop: Use a while True loop to create an infinite loop. This loop will keep running indefinitely, publishing temperature data to the MQTT broker in regular intervals.
5.	Generate a random temperature value: Generate a random floating-point number between 20.0 (inclusive) and 21.0 (exclusive) using the uniform(20.0, 21.0) function. Assign this value to the randNumber variable.
6.	Publish the temperature value: Use the publish() method of the MQTT client to publish the random temperature value to the MQTT topic "TEMPERATURE". The topic name is specified as the first argument, and the temperature value is sent as the payload.
7.	Print the published message: Display a message indicating that the temperature value has been published to the "TEMPERATURE" topic. The message includes the random temperature value converted to a string.
8.	Add a delay: Use the time.sleep(1) function to introduce a 1-second delay before the loop repeats. This means that the temperature value will be published to the MQTT broker every second.
9.	Repeat the loop: The loop continues to generate and publish random temperature values indefinitely. As a result, the MQTT broker will receive temperature data at regular intervals, simulating the behavior of a temperature sensor publishing data to the broker.
In summary, the code sets up an MQTT client, connects it to an MQTT broker, and then enters an infinite loop that generates random temperature data and publishes it to the "TEMPERATURE" topic every second. This simulates a temperature sensor sending temperature readings to the MQTT broker.

```python
import paho.mqtt.client as mqtt
from random import randrange, uniform
import time
```

`import paho.mqtt.client as mqtt:`

This line imports the paho.mqtt.client module, which allows the Python script to interact with the MQTT (Message Queuing Telemetry Transport) protocol.

`from random import randrange,uniform:`

This line imports the randrange and uniform functions from the random module. These functions are used to generate random numbers.

`import time:`

This line imports the time module, which is used to introduce delays in the code execution.

```python
mqttBroker="mqtt.eclipseprojects.io"
#mqttPort=1883
```

This line sets the MQTT broker address to `"mqtt.eclipseprojects.io"`. The broker is responsible for handling the messages between clients. The popular broker port number is 1883

```python
client=mqtt.Client("Temperature_Inside")
client.connect(mqttBroker)
```

This line creates an MQTT client instance with the name `"Temperature_Inside"`. The client will be used to connect to the MQTT broker and publish messages.

```python
while True:
  randNumber=uniform(20.0,21.0)
  client.publish("TEMPERATURE",randNumber)
  print("Just Published " +str(randNumber)+ " to topic TEMPERATURE")
  time.sleep(1)
```

`while True: `

This starts an infinite loop that will keep running until the program is manually interrupted or terminated.

`randNumber = uniform(20.0, 21.0): `

This line generates a random floating-point number between 20.0 (inclusive) and 21.0 (exclusive) and assigns it to the variable randNumber.

`client.publish("TEMPERATURE", randNumber):`

This line publishes the random number (randNumber) to the MQTT topic "TEMPERATURE" using the MQTT client. The temperature value is sent as a payload to the topic.

`print("Just published " + str(randNumber) + " to Topic TEMPERATURE"):`

This line prints a message indicating that the temperature value has been published to the "TEMPERATURE" topic. The str() function is used to convert the random number to a string for concatenation with the message.

`time.sleep(1):`

This line introduces a 1-second delay before the loop iterates again. This means that the temperature value will be published every second.
